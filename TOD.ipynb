```python
df1=df.drop(columns=['Qchat-10-Score'])
df1
```

```python
label_enc = LabelEncoder()
df1["Ethnicity"] = label_enc.fit_transform(df["Ethnicity"])
df1["Who completed the test"] = label_enc.fit_transform(df["Who completed the test"])
```

```python
df1
```

```python
le = LabelEncoder()
columns = [ 'Family_mem_with_ASD', 'Class/ASD Traits ', 'Sex', 'Jaundice']
for col in columns:
    df1[col] = le.fit_transform(df1[col])
df1.dtypes
```

```python
df1
```

```python
df2=df1.drop(columns=['Case_No'])
df2
```

```python
df2.corr()
```

```python
df3=df2.drop(columns=['Who completed the test'])
df3
```

```python
X=df3.iloc[:,:-1]
y=df3.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi=LogisticRegression(fit_intercept=True,max_iter=10000)
logi.fit(X_train,y_train)
```

```python
y_pred_logi=logi.predict(X_test)
```

```python

print(accuracy_score(y_pred_logi,y_test))
```

```python
confusion_matrix(y_test,y_pred_logi)
```

```python
import numpy as np
fe=np.array([[0,0,1,0,0,0,0,0,1,0,48,0,2,0,0]])
print(logi.predict(fe))
```

```python
import joblib
joblib.dump(logi, 'autism_model.pkl')
```
